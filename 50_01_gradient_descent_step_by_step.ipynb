```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_targets=1,noise=80 , random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# lets apply ols
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Lets apply Gradient Descent assuming slope is m = 78.35
# and lets assume the starting value for intercetpt b = 0
y_pred = ((78.35*X)+0)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red' , label='OLS')
plt.plot(X,y_pred,color='green' , label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
lr = 0.1

step_size  = loss_slope*lr
step_size
```

```python
# calculating new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red' , label='OLS')
plt.plot(X,y_pred1 , color = 'Blue',label='b = {}'.format(b) )
plt.plot(X,y_pred,color='green' , label='b = 0')

plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size  = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red' , label='OLS')
plt.plot(X,y_pred2 , color = 'black',label='b = {}'.format(b) )
plt.plot(X,y_pred1 , color = 'Blue',label='b = {}'.format(b) )
plt.plot(X,y_pred,color='green' , label='b = 0')

plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red' , label='OLS')
plt.plot(X,y_pred3 , color = 'yellow',label='b = {}'.format(b) )
plt.plot(X,y_pred2 , color = 'black',label='b = {}'.format(b) )
plt.plot(X,y_pred1 , color = 'Blue',label='b = {}'.format(b) )
plt.plot(X,y_pred,color='green' , label='b = 0')

plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.01

epochs = 100

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    step_size  = loss_slope*lr
    b = b - step_size

    y_pred = ((78.35*X)+b).reshape(4)
    plt.plot(X,y_pred)


plt.scatter(X,y)
```
