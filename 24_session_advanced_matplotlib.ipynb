```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c = iris['Species'],cmap = 'winter',alpha = 1.0)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.colorbar()
```

```python
# cmap and alpha
```

### Plot size

```python
plt.figure(figsize = (15,7))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c = iris['Species'],cmap = 'winter',alpha = 1.0)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.colorbar()
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)

yy
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z, cmap='viridis')

fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z, cmap='viridis')

fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z, cmap='viridis')

fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z, cmap='viridis')

fig.colorbar(p)
```

### Heatmap

```python
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')

grid
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(list(np.arange(0,6)))
plt.colorbar()
```
